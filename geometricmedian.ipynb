```python
#imports 
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
import seaborn as sns
from matplotlib.colors import LogNorm, Normalize
from matplotlib.ticker import MaxNLocator
from random import randint
from math import ceil
from scipy.spatial import ConvexHull

# functions
def dist(p1, p2):
	return sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)


def outofrange(m):
    if (m[0] < mnx or m[0] > mxx or m[1] < mny or m[1] > mxy):
        return True

def aboveline(ln, pt):
    # ln is in the form (m, c, >=?, <=?), pt is in the form (x, y);
    if (pt[0] < ln[2] or pt[0] > ln[3]):
        return 0
        
    if pt[1] >= pt[0] * ln[0] + ln[1]:
        return 1
    else:
        return 0

def countlines(pt):
    abv = 0
    for ln in lines:
        abv += aboveline(ln, pt);
    return abv

def inshape(pt):
    abv = countlines(pt)
    if(abv == 1):
        return True
    else:
        return False;

def closeenough(ln, pt):
    if (pt[0] < ln[2] or pt[0] > ln[3]):
        return 0
    if pt[0] * ln[0] + ln[1] - pt[1] < 1:
        return 1
    else:
        return 0

def findedge(pt):
    for (ind, ln) in enumerate(lines):
        if (closeenough(ln, pt)):
            return "ON EDGE " + str(ind)
    return "NOT ON EDGE"


for itr in range(10):
    # generate points
    # temp = [(0, 87), (50, 173), (150, 173),(200,87),(50,0),(150,0)]
    temp = [(0,0)]
    points = []
    
    temp.append((randint(0, 2000), randint(0, 2000)))
    temp.append((randint(0, 2000), 0))


    # for pt in range(1):
    #     temp.append((randint(0, 2000), randint(0, 2000)))

    hull = ConvexHull(temp)

    for vert in hull.vertices:
        points.append(temp[vert])
    points = np.array(points)
    n = len(points)
    mnx = np.min(points[:, 0])
    mxx = np.max(points[:, 0])
    mny = np.min(points[:, 1])
    mxy = np.max(points[:, 1])
    
    lines = []
    for i in range(0, n):
        j = (i + 1) % n;
        if (points[i][0] - points[j][0] == 0):
            continue;
        m = (points[i][1] - points[j][1])/(points[i][0] - points[j][0])
        c = points[i][1] - points[i][0] * m;
        lines.append((m, c, min(points[i][0], points[j][0]), max(points[i][0], points[j][0])));
    
    print(lines)
    print(mnx, mny, mxx, mxy)
    
    # assert(n==len(lines)) 
    # assert might be false if the points are right above each other
    
    maxpoint = (0, 0)
    maxvalue = 0
    minpoint = (0, 0)
    minvalue = 1e10
    plotx = []
    ploty = []
    values = []
    plot = [[0] * (mxx + 1)for i in range (mxy + 1)]
    
    for x in range(mnx, mxx):
        for y in range(mny, mxy):
            cur = (x, y) # this is the point we are checking
            if not inshape(cur):
                continue
            td = 1
            for pt in points:
                td *= dist(pt, cur)
            if td > maxvalue: 
                maxpoint = cur
                maxvalue = td
            if td < minvalue:
                minpoint = cur
                minvalue = td
            values.append(td)
            plot[y][x] = td
    
    findedge(maxpoint)
    values = np.array(values)
    print(values)
    np.sort(values)
    plot = np.array(plot)
    cutoff = values[round(len(values) / 1000)]
    plot = np.where(plot < cutoff, cutoff, plot)
    
    # highlight maxpoint
    scalex = 1 
    scaley = round((mxy - mny) / (mxx - mnx))
    pointsize = ceil((mxx - mnx)/125)
    for x in range(max(mnx, maxpoint[0] - pointsize), min(mxx, maxpoint[0] + pointsize)):
        for y in range(max(mny, maxpoint[1] - pointsize * scaley), min(mxy, maxpoint[1] + pointsize *scaley)):
            plot[y][x] = maxvalue * 1.5
            
    ttl = str(n) + "-gon"
    cap = "Polygon vertices " 
    for pt in points:
        cap += "(" + str(pt[0]) + ", " + str(pt[1]) + ") "
    ttl += ", max point (" + str(maxpoint[0]) + ", " + str(maxpoint[1]) + ")"
    print(points)
    print("hohoh", maxpoint)
    
    ax = plt.axes()
    hm = sns.heatmap(data=plot)
    hm = hm.invert_yaxis()
    
    SMALL_SIZE = 7
    MEDIUM_SIZE = 8
    BIGGER_SIZE = 12
    
    plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
    
    ax.set_title(ttl)
    plt.figtext(0.5, 0.01, cap, wrap=True, horizontalalignment='center', fontsize=12)
    plt.tight_layout()
    filename = "polyheatmap" + str(itr) +".png"
    plt.savefig(fname=filename, format="png", pad_inches=0.1)
    plt.close()
    
    plt.figure(dpi=400)
    plt.axis('off')
    ax = plt.gca()

    plt.plot(points[:, 0], points[:, 1], 'ko-')
    plt.plot(maxpoint[0], maxpoint[1], 'kx')
    plt.plot([points[n-1][0], points[0][0]], [points[n-1][1], points[0][1]], "ko-")
    ax.set_aspect('equal', adjustable='box')
    
    for pt in points:
        print("annotating pt")
        os = 35
        if (pt[1] > mxy - 50):
            os = -35
        plt.annotate(f"({pt[0]}, {pt[1]})", (pt[0],pt[1]+os)) 
    plt.annotate(f"({maxpoint[0]}, {maxpoint[1]})", (maxpoint[0], maxpoint[1] + 25))


    # for pt in points:
    #     txt = "(" + str(pt[0]) + ", " + str(pt[1]) + ")"
    #     plt.annotate(txt, (pt[0], pt[1]))

        
    filename = "polylines" + str(itr) +".png"

    plt.savefig(fname=filename, format="png", pad_inches=-0.1)
    plt.show()
    plt.close()
```
